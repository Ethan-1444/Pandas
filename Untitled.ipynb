```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# import libaries needed to graph the model
```

```python
my_csv=pd.read_csv('flat-ui__data-Fri Apr 04 2025.csv')
my_csv
# using pandas to read the csv file
```

```python
ph=my_csv.query('Country == "Philippines"')
ph
# Finding Philippines in the database
```

```python
ph_1976=ph.query('Year>=1976')
ph_1976
# setting the initial year=1976
```

```python
year=ph_1976['Year']
year
#Selecting the year/ independent variable so it can be used in the x axis
```

```python
cpi=ph_1976["CPI"].round(2)
cpi
#Indexing the cpi/dependent variable to be used in the y-axis
```

```python
x=year
y=cpi
# Assigning the x and y to the year and cpi respectively
plt.plot(x,y)
plt.scatter(x, y, label='Data Points')
# Ploting the points of x and y
plt.xlabel('Year')
plt.ylabel('CPI')
# Putting label of x and y as Year and CPI
slope, intercept = np.polyfit(x, y,1)
# Using polyfitting where the highest degree is 1
# The polyfit method gives use the equation of the line which is y=mx+b
plt.title(f'Annual CPI of the Philippines (y={round(slope,2)}x+{round(intercept,2)})')
#Making the title of the model
regression_line = slope * x + intercept
plt.plot(x, regression_line, color='red', label='Regression Line')
plt.grid(True)
# Graphing the linear function
print(plt.show())
# Annual change of 0.3
```
