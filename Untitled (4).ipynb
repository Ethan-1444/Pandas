```python
x=years
y=value
plt.plot(x,y)
plt.scatter(x,y,label="Data points")
plt.xlabel("Year")
plt.ylabel("PPP")
slope,intercept=np.polyfit(x,y,1)
plt.title(f'Annual Purchasing Power Parity of the Philippines(y={round(slope,2)}x+{round(intercept,2)}')
linear_regression=slope*x+intercept
plt.plot(x,linear_regression,color="purple", label="Line")
plt.grid(True)
print(plt.show())
```
