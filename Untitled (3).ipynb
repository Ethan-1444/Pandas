```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
m=95
c=100000
```

```python
x=np.linspace(1000,2000,5)
```

```python
y=m*x-c
```

```python
plt.plot(x,y)
plt.xlabel('Cups sold')
plt.ylabel('Profit')
plt.title('Linear model for the Cup Business')
plt.grid(True)
plt.show()
```
